```python
#Q1
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

tesla = yf.Ticker("TSLA")

tesla_data = tesla.history(period="max")

tesla_data = tesla_data.reset_index(drop=True)

tesla_data = tesla_data.head()
```

```python
#Q2
tesla_revenue = tesla_data.tail()
```

```python
#Q3
gme_data = yf.Ticker("GME")

gme_data = gme_data.history(period="max")

gme_data = gme_data.reset_index(drop=True)

gme_data=gme_data.head()
```

```python
#Q4
gme_revenue = gme_data.tail()
```

```python
#Q5
def make_graph(data, title):
    plt.figure(figsize=(10, 6))
    plt.plot(data['Close'], label='Close Price')
    plt.xlabel('id')
    plt.ylabel('Close Price')
    plt.title(title)
    plt.legend()
    plt.show()

make_graph(tesla_data, "Tesla Stock Data")
```

```python
#Q6
make_graph(gme_data, "GME Stock Data")
```
